```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\chandana\Music\Desktop\practice\day12\housing.csv")
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='median_income',
                y='median_house_value',
                data=df)

plt.title("Median Income vs Median House Value")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")

plt.show()
```

```python
correlation = df[['median_income', 'median_house_value']].corr()

print(correlation)
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='ocean_proximity',
            y='median_house_value',
            data=df)

plt.title("House Value by Ocean Proximity")
plt.xticks(rotation=45)

plt.show()
```
