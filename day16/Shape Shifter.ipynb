```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Normal Distribution (Heights)
heights = np.random.normal(loc=165, scale=10, size=1000)

# Right-Skewed (Income)
income = np.random.exponential(scale=50000, size=1000)

# Left-Skewed (Easy Exam Scores)
scores = 100 - np.random.exponential(scale=10, size=1000)
scores = np.clip(scores, 0, 100)  # keep between 0 and 100
```

```python
def plot_distribution(data, title):
    plt.figure()
    sns.histplot(data, kde=True)
    plt.title(title)
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()
    
    print("Mean:", np.mean(data))
    print("Median:", np.median(data))
    print("-"*40)
```

```python
plot_distribution(heights, "Normal Distribution - Heights")
plot_distribution(income, "Right-Skewed Distribution - Income")
plot_distribution(scores, "Left-Skewed Distribution - Scores")
```
