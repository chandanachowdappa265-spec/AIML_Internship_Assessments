```python
# Independent Events

# Probability of Heads
P_heads = 1/2

# Probability of rolling a 6
P_six = 1/6

# Since independent:
P_both = P_heads * P_six

print("P(Heads AND 6) =", P_both)
```

```python
# Dependent Events

# First red
P_first_red = 5/10

# After picking one red, remaining:
# 4 Red out of 9 total
P_second_red = 4/9

P_both_red = P_first_red * P_second_red

print("P(Both are Red) =", P_both_red)
```
